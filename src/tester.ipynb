```python
import folium
geo_center_of_US = [45.0, -100.0]
map = folium.Map(location=geo_center_of_US,zoom_start=3,tiles='Stamen Terrain')
map
```

```python
places = r
ind_to_color = {
        'H' : 'red',
        'W' : 'pink',
        'N' : 'green',
        'C' : 'light blue',
        'F' : 'blue'
        }
for plist in places.values():
    lat = float(plist['lat'])
    long = float(plist['long'])
    ind = plist['indicator']
    name = plist['name']+ ', ' + plist['state']
    color = ind_to_color[ind]
    marker = folium.Marker(  #features.CircleMarker(
        location=[lat,long],
        popup=name,
        icon=folium.Icon(color=color,icon='cloud')
        #radius=1,
        #color=color,
        #fill=True,
        #fill_color=color)
    )
    marker.add_to(map)
map
```
