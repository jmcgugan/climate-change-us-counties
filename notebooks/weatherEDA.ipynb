```python
def get_seq(location, dataset='GSOM', startyear=1800, endyear=2017):
    result = pd.DataFrame()
    endpoint = "https://www.ncdc.noaa.gov/cdo-web/api/v2/data?"
    payload = {
        'datasetid': dataset,
        'locationid': location, 
        'startdate': '1800-01-01',
        'enddate' : '1800-12-31',
        'offset' : 1,
        'limit' : 1000}
    for year in range(startyear, endyear+1):
        payload['startdate'] = str(year) + '-01-01'
        payload['enddate'] = str(year) + '-12-31'
        response = requests.get(endpoint, headers=headers, params=payload)
        if response.status_code != 200:
            raise ValueError('Response Code from NOAA not 200')
        batch = pd.DataFrame.from_records(response.json()['results'])
        result = result.append(batch)
    return result
        
```

```python
def get_seq(location, dataset='GSOM', startyear=1800, endyear=2017):
    batchlimit = 1000
    result = pd.DataFrame()
    endpoint = "https://www.ncdc.noaa.gov/cdo-web/api/v2/data?"
    payload = {
        'datasetid': dataset,
        'locationid': location, 
        'startdate': '1800-01-01',
        'enddate' : '1800-12-31',
        'offset' : 1,
        'limit' : batchlimit}
    for year in range(startyear, endyear+1):
        payload['startdate'] = str(year) + '-01-01'
        payload['enddate'] = str(year) + '-12-31'
        offset = 1
        payload['offset'] = offset
        count = batchlimit + 1 #just to get started; first call will tell us how many
        while offset <= count:
            print('Year: ' + str(year) + ' Fetching at offset' + str(offset) + ' Count: ' + str(count))
            response = requests.get(endpoint, headers=headers, params=payload)
            if response.status_code != 200:
                raise Exception('Response Code from NOAA not 200')
            count = int(response.json()['metadata']['resultset']['count']) #true count
            batch = pd.DataFrame.from_records(response.json()['results'])
            result = result.append(batch)
            offset = offset + batchlimit
    return result
```
